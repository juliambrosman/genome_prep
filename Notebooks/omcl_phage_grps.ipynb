```python
phage_codes = "../../OrthoMCL/phage_v_phage/phage_taxon_codes.txt"
names = {}
get_name = lambda i: i.split("\t")[0].split("/")[-1].replace("fasta","")


with open(phage_codes) as pc:
    for i, p in enumerate(pc):
        names[p.split("\t")[1].strip()] = get_name(p)
```

```python
phage_pcs = defaultdict(lambda:set())
```

```python
groups = "../data/groups.txt"
with open(groups) as infile:
    for i, l in enumerate(infile):
        pc = l.split(":")[0]
        lst = l.split(":")[1]
        for p in lst.split():
            phage_pcs[names[p.split("|")[0]]].add(pc)
```

```python
shared = defaultdict(lambda: [])

for p in phage_pcs:
    for n in phage_pcs:
        if phage_pcs[p] == phage_pcs[n] and p != n:
            shared[p].append(n)    
```

```python
out = "../data/nvp_100pct_shared_clusts.txt"

with open(out, "w") as oh:
    for p in shared:
        print(p, ",".join(shared[p]), sep="\t", file=oh)
```
